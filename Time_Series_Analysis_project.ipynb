# Run an Autoregressive Model (Linear Regression) on the remainder

```python
# Drop missing values
df_temp.dropna(inplace=True)
df_temp.head()
```

```python
X_all_features  = x.copy()
X_all_features['remainder']= df_temp['remainder']
X_all_features['lag1'] = df_temp['lag1']
X_all_features['lag2'] = df_temp['lag2']
X_all_features['lag3'] = df_temp['lag3']
X_all_features['lag4'] = df_temp['lag4']
X_all_features
```

```python
X_all_features.dropna(inplace=True)
X_all_features
```

```python
# Assign X and y
X = X_all_features.drop('remainder',axis=1)
y = df_temp['TG']
```

```python
# Create and fit the model
ar_model = LinearRegression()
ar_model.fit(X, y)
```

```python
# Create predictions
df_temp['ar_predictions'] = ar_model.predict(X)
```

```python
# Plot the original remainder and the prediction
rcParams["figure.figsize"] = [15.4, 4.8]
plt.plot(df_temp[['TG', 'ar_predictions']])
```

```python
# Inspect the residual after modelling the remainder
rcParams["figure.figsize"] = [15.4, 4.8]
plt.plot(df_temp.remainder - df_temp.ar_predictions)
```

```python
# custom plot function

def plot_timeseries(df, title="", ylim=False):
    """
    Custom plotting function for plotting the flights dataset
    
    Parameters
    ----------
    df : pd.DataFrame
        The data to plot.
    title : str
        The title of the plot
    ylim : bool
        Whether to fix the minimum value of y; defalut is True
    
    Returns
    -------
    Plots the data
    
    """
    
    df.plot(linestyle='-', marker='.')
    plt.title(title)
    if ylim:
        plt.ylim(ymin=0)
    plt.ylabel("TG")
    plt.show()
```
