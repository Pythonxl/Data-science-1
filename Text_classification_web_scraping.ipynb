```python
names_2 = []
links_2 = []
for link_2 in links_list_2:
    if link_2['href'].startswith('/lyric'):
        names_2.append(link_2.text)
        links_2.append('https://www.lyrics.com' + link_2['href'])
```

```python
header={'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 6.1; tr-TR) AppleWebKit/533.20.25 (KHTML, like Gecko) Version/5.0.4 Safari/533.20.27'}
```

```python
lyric_pages_1 = []
for song_link_1 in links_1[:50]:
    response_1 = requests.get(song_link_1, headers=header)  # call the web server and get the web page
    time.sleep(np.random.uniform(1,2,1)[0])  # wait for some time to not be too fast 
    lyric_pages_1.append(response_1.text)  # save the whole web pages in a list
```

```python
lyric_pages_2 = []
for song_link_2 in links_2[:50]:
    response_2= requests.get(song_link_2, headers=header)  # call the web server and get the web page
    time.sleep(np.random.uniform(1,2,1)[0])  # wait for some time to not be too fast 
    lyric_pages_2.append(response_2.text)  # save the whole web pages in a list
```

Extract the actual lyrics from the lyrics web page, that we downloaded

```python
lyrics_1 = []
for lyric_page_1 in lyric_pages_1:
    my_soup_1 = BeautifulSoup(lyric_page_1, parser='html-parser')
    lyrics_1.append(my_soup_1.find('pre', attrs={'id': 'lyric-body-text'}).text)
```

```python
lyrics_2 = []
for lyric_page_2 in lyric_pages_2:
    my_soup_2 = BeautifulSoup(lyric_page_2, parser='html-parser')
    lyrics_2.append(my_soup_2.find('pre', attrs={'id': 'lyric-body-text'}).text)
```

Save the actual song lyrics in text files

```python
for i, lyric in enumerate(lyrics_1):
    with open(f"EP_{names_1[i].replace('/', '')}.txt", 'w') as lyrics_file_1:
        lyrics_file_1.write(lyric)
```

```python
for i, lyric in enumerate(lyrics_2):
    with open(f"MJ_{names_2[i].replace('/', '')}.txt", 'w') as lyrics_file_2:
        lyrics_file_2.write(lyric)
```
